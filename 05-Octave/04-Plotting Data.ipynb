```octave
t = [0:0.01:0.98];
```

```octave
y1 = sin(2*pi*4*t);
```

```octave
plot(t, y1)
```

```octave
y2 = cos(2*pi*4*t);
```

```octave
plot(t, y2)
```

```octave
plot(t, y1)
hold on
plot(t, y2, 'r')
xlabel('time')
ylabel('value')
legend('sin', 'cos')
title('my plot')

print -dpng 'myPlot.png'
```

```octave
subplot(1,2,1)
plot(t, y1)
subplot(1,2,2)
plot(t, y2)
```
